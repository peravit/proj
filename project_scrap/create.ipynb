```python
import re
import string
import os
import pandas as pd
import numpy as np

from collections import Counter

from pythainlp.tokenize import word_tokenize
```

```python
def getNgramsFromSentence(content, n):
    output = []
    for i in range(len(content)-n+1):
        output.append(content[i:i+n])
    return output

```

```python
df = pd.read_csv('twitter.csv')
```

```python
df
```

```python
del_word = [' ','\n','"','\u200b']
sen = []

for i in df['text']:

    t = '< ' + str(i) + ' >'
    t2 = word_tokenize(t)
    
    for j in t2:
        if j not in del_word:
            sen.append(j)
            
```

```python
sen
```

```python
# 1 - grams
n = 1


grams = getNgramsFromSentence(sen, n)

values, counts = np.unique(grams, return_counts=True)

dfc = pd.DataFrame({'words': values , 'count': counts})
dfc = dfc.sort_values(by='count', ascending=False , axis = 0)
dfc = dfc.reset_index()

dfc.to_csv('1/' + str(n) + '-grams.csv',encoding="utf-8")
```
