```python
import nltk.tag, nltk.data
import pickle
with open('thaipos.pickle', 'rb') as handle:
    model = pickle.load(handle)
```

```python
sentence =['บทความ หมายถึงงานเขียนที่เผยแพร่ในสื่อสิ่งพิมพ์หรือสื่ออิเล็กทรอนิกส์',
           'มีจุดประสงค์เพื่อเผยแพร่ข่าวสาร ผลการวิจัย เผยแพร่ความรู้ การวิเคราะห์ทางการศึกษา การวิพากษ์วิจารณ์ เป็นต้น',
           'โดยปกติบทความหนึ่งบทความจะพูดถึงเรื่องใดเรื่องหนึ่งเป็นประเด็นหลักเพียงเรื่องเดียว',
           'หัวเรื่องหรือพาดหัว (headline) เป็นข้อความที่อยู่บนสุดของบทความแสดงถึงประเด็นหลักของบทความ',
           'หัวเรื่องมักถูกออกแบบให้โดดเด่นสะดุดตา และใช้คำที่เลือกสรรมาให้เกี่ยวข้องกับหัวเรื่อง']
```

```python
sen_to_tag = []
for i in sentence:
    text=word_tokenize(i)
    tagger = nltk.tag.UnigramTagger(model=model)
    pos = tagger.tag(text)
    sen_to_tag.append(pos)
    
for i in range(len(sen_to_tag)):
    sen_to_tag[i].insert(0,('<s>','<s>'))
    sen_to_tag[i].insert(len(sen_to_tag[i]),('</s>','</s>'))
```

```python
sen_to_tag
```

```python
word_tag_name = []
word_tag_count = []

for i in sen_to_tag:
    for j in i:
        if j not in word_tag_name:
            word_tag_name.append(j)
            word_tag_count.append(1)
        else:
            word_tag_count[word_tag_name.index(j)] = word_tag_count[word_tag_name.index(j)] + 1
```

```python
tag_name = []
word_name = []
word_name_count = []

for i in sen_to_tag:
    for j in i:
        if j[1] not in tag_name:
            tag_name.append(j[1])
            word_name.append([j[0]])
            word_name_count.append([1])
        else:
            if j[0] not in word_name[tag_name.index(j[1])]:
                word_name[tag_name.index(j[1])].append(j[0])
                word_name_count[tag_name.index(j[1])].append(1)
            else:
                word_name_count[tag_name.index(j[1])][word_name[tag_name.index(j[1])].index(j[0])] = word_name_count[tag_name.index(j[1])][word_name[tag_name.index(j[1])].index(j[0])] + 1
```

```python
sen_only_tag = []

n = 0
for i in sen_to_tag:
    sen_only_tag.append([])
    for j in i:
        sen_only_tag[n].append(j[1])
    n += 1
```

```python
bigram_tag_name = []
bigram_tag_count = []

for i in sen_only_tag:
    for j in range(len(i) - 1):
        t = (i[j],i[j + 1])
        if t not in bigram_tag_name:
            bigram_tag_name.append(t)
            bigram_tag_count.append(1)
        else:
            bigram_tag_count[bigram_tag_name.index(t)] = bigram_tag_count[bigram_tag_name.index(t)] + 1
            
```

```python
sen = ['บทความ หมายถึงงานเขียนที่เผยแพร่ในสื่ออิเล็กทรอนิกส์',
       'หัวเรื่องและเนื้อเรื่องควรสัมพันธ์กัน',
       'มีจุดประสงค์เพื่อเผยแพร่ข่าวสาร การวิเคราะห์ทางการศึกษา เผยแพร่ความรู้ ผลการวิจัย การวิพากษ์วิจารณ์ เป็นต้น',
       'หัวเรื่องเป็นข้อความที่อยู่บนสุดของบทความแสดงถึงประเด็นหลักของบทความ',
       'บทความ หมายถึงงานเขียนที่เผยแพร่ในสื่อสิ่งพิมพ์หรือสื่ออิเล็กทรอนิกส์']
```

```python
sen_ex = []
for i in sen:
    text=word_tokenize(i)
    tagger = nltk.tag.UnigramTagger(model=model)
    pos = tagger.tag(text)
    sen_ex.append(pos)
    
for i in range(len(sen_ex)):
    sen_ex[i].insert(0,('<s>','<s>'))
    sen_ex[i].insert(len(sen_ex[i]),('</s>','</s>'))
    
sen_ex
```

```python
ex_prob_transition = []
ex_prob_emission = []
```

```python
for i in range(len(sen_ex)):
    ex_prob_transition.append([])
    for j in range(len(sen_ex[i]) - 1):
      try:
        ex_prob_transition[i].append(transition_prob.loc[sen_ex[i][j][1],sen_ex[i][j + 1][1]])
      except:
        ex_prob_transition[i].append(0)
```

```python
for i in range(len(sen_ex)):
    ex_prob_emission.append([])
    for j in range(len(sen_ex[i]) - 2):
        try:
            ex_prob_emission[i].append(emission_prob.loc[sen_ex[i][j + 1][0],sen_ex[i][j + 1][1]])
        except:
            ex_prob_emission[i].append(0)
```

```python
ex_prob_emission
```

```python
ex_prob_transition
```

```python
prob = []

for i in range(len(ex_prob_transition)):
    p = 1
    for j in ex_prob_emission[i] + ex_prob_transition[i]:
        p = p * j
    prob.append(p)
```

```python
for i in range(len(sen)):
    print('Sentence    : ',sen[i])
    print('Probability : ',prob[i])
    print('-'*50)
```
